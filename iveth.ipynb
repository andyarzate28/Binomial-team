## Ejercicio 4
Pag 165

Seguros. De acuerdo con la Asociación Mexicana de Instituciones de Seguros (amis), en 2010 sólo
47% de los automóviles contaban con un seguro. Si se realizara una revisión a 18 automóviles:

a) ¿Cómo sería su distribución? Determina la distribución de probabilidad, traza la gráfica
correspondiente e indica el tipo de sesgo que presenta.

Es una distribución binomial porque:

Hay un número fijo de ensayos

Cada auto puede tener o no tener seguro (dos posibles resultados).



```python
#Trazar Grafica
n=18
p=0.47
k=list(range(n+1))
y=binom.pmf(k,n,p)
plt.bar(k, y)
plt.xlabel("Valores de k")
plt.ylabel("Probabilidades")
plt.title("Distribucion binomial")
plt.show()
```

La grafica no esta sesgada 

c) ¿Cuál es la probabilidad de que 13 o más autos tengan seguro?

```python
n=18
p=0.47
k=range(13, n+1)
y=binom.pmf(k,n,p)
print(sum(y))
```

## Ejercicio 5
Pag 165

Sexualidad. Con base en la Encuesta Nacional de Juventud 2005, 90% de los adolescentes entre
15 y 19 años afirmó que tuvo su primera relación sexual antes de los 18 años. Si en un grupo
hay 20 adolescentes entre 15 y 19 años, entonces:

```python
#Trazar Grafica
n=20
p=0.90
k=list(range(n+1))
y=binom.pmf(k,n,p)
plt.bar(k, y)
plt.xlabel("Valores de k")
plt.ylabel("Probabilidades")
plt.title("Distribucion binomial")
plt.show()
```

La grafica esta sesgada a la derecha

 a) ¿Cuántos esperarías que contestaran que tuvieron su primera relación sexual antes de los
18 años?

```python
n=20
p=0.90
resultado=n*p
print(resultado)
```

b) ¿Cuál es la probabilidad de que 12 de ellos digan que tuvieron su primera relación sexual antes de los 18 años?

```python
n=20
p=0.90
k=[ 0,1,2,3,4,6,7,8,9,10,11,12]
y=binom.pmf(k,n,p)

print (sum(y))
```

c) ¿Cuál es la probabilidad de que más de 17 digan que tuvieron su primera relación sexual
antes de los 18 años?

```python
n=20
p=0.90
k=range(17, n+1)
y=binom.pmf(k,n,p)
print(sum(y))
```

a) ¿Cuál es la probabilidad de que 10 de ellas consuman alguna de estas drogas?

b) ¿Cuál es la probabilidad de que al menos 20 de ellas consuman alguna de estas drogas?

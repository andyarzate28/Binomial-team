Es una distribuci칩n binomial porque:

```python
#Trazar Grafica
n=18
p=0.47
k=list(range(n+1))
y=binom.pmf(k,n,p)
plt.bar(k, y)
plt.xlabel("Valores de k")
plt.ylabel("Probabilidades")
plt.title("Distribucion binomial")
plt.show()
```

# Ejercicio 7. Lanzamiento de monedas
 Página 180
 Libro: Schaum's Outline of Introduction to Probability and Statistics
 Autores: SEYMOUR LIPSCHUTZ, JOHN J. SCHILLER
 Problema: 
 Se lanza una moneda justa 10 veces. ¿Cuál es la probabilidad de obtener exactamente 4 caras?



```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parámetros del ejercicio
n = 10          # número de ensayos
p = 0.5         # probabilidad de éxito (cara)
k = 4           # número de éxitos deseado

# Valores posibles de X
x = np.arange(0, n + 1)

# Función de masa de probabilidad binomial
y = binom.pmf(x, n, p)

# Graficar la distribución binomial
plt.figure(figsize=(8, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.bar(k, binom.pmf(k, n, p), color="orange", edgecolor="black", label=f"P(X = {k})")

# Personalización del gráfico
plt.title("Distribución Binomial: Lanzamiento de moneda (n=10, p=0.5)")
plt.xlabel("Número de caras (x)")
plt.ylabel("Probabilidad P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

```python
# Ejercicio 8. Piezas defectuosas
# Página 189-190
# Libro: Schaum's Outline of Introduction to Probability and Statistics
# Autores: SEYMOUR LIPSCHUTZ, JOHN J. SCHILLER
# Problema: 
# De una línea de producción, se sabe que cada pieza tiene un 2 % de probabilidad de estar defectuosa. Si se inspeccionan 50 piezas, ¿cuál es la probabilidad de encontrar exactamente 2 defectuosas?
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parámetros del ejercicio 3
n = 50          # número de piezas
p = 0.02        # probabilidad de que una pieza esté defectuosa
k = 2           # número de piezas defectuosas

# Valores posibles de X
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

# Gráfica
plt.figure(figsize=(10, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.bar(k, binom.pmf(k, n, p), color="orange", edgecolor="black", label=f"P(X = {k})")

plt.title("Distribución Binomial: Piezas defectuosas (n=50, p=0.02)")
plt.xlabel("Número de piezas defectuosas (x)")
plt.ylabel("Probabilidad P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
# Ejercicio 9. Encuesta Politica
# Página 204-206
# Libro: Schaum's Outline of Introduction to Probability and Statistics
# Autores: SEYMOUR LIPSCHUTZ, JOHN J. SCHILLER
# Problema: 
# En una encuesta política, la probabilidad de que una persona vote por el candidato A es 0.6. Si se preguntan 20 personas, ¿qué probabilidad hay de que exactamente 15 elijan al candidato A?
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parámetros del ejercicio 2
n = 20          # número de personas encuestadas
p = 0.6         # probabilidad de que voten por A
k = 15          # número de personas que votan por A

# Valores posibles de X
x = np.arange(0, n + 1)

# Función de masa de probabilidad binomial
y = binom.pmf(x, n, p)

# Gráfica
plt.figure(figsize=(8, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.bar(k, binom.pmf(k, n, p), color="orange", edgecolor="black", label=f"P(X = {k})")

# Etiquetas y detalles
plt.title("Distribución Binomial: Encuesta política (n=20, p=0.6)")
plt.xlabel("Personas que votan por A (x)")
plt.ylabel("Probabilidad P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

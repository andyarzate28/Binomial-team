# Ejercicio 7. Lanzamiento de monedas
 P√°gina 180
 Libro: Schaum's Outline of Introduction to Probability and Statistics
 Autores: SEYMOUR LIPSCHUTZ, JOHN J. SCHILLER
 Problema: 
 a) Determina la distribuci√≥n de probabilidad y traza la gr√°fica correspondiente.
Es una distribuci√≥n binomial porque:
Hay un n√∫mero fijo de ensayos: 
ùëõ=10
n=10 lanzamientos de moneda.
Cada lanzamiento tiene dos posibles resultados: cara o cruz.
La probabilidad de obtener una cara (√©xito) es constante: 
ùëù=0.5
p=0.5.
Los lanzamientos son independientes entre s√≠.



```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Par√°metros del ejercicio
n = 10          # n√∫mero de ensayos
p = 0.5         # probabilidad de √©xito (cara)
k = 4           # n√∫mero de √©xitos deseado

# Valores posibles de X
x = np.arange(0, n + 1)

# Funci√≥n de masa de probabilidad binomial
y = binom.pmf(x, n, p)

# Graficar la distribuci√≥n binomial
plt.figure(figsize=(8, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.bar(k, binom.pmf(k, n, p), color="orange", edgecolor="black", label=f"P(X = {k})")

# Personalizaci√≥n del gr√°fico
plt.title("Distribuci√≥n Binomial: Lanzamiento de moneda (n=10, p=0.5)")
plt.xlabel("N√∫mero de caras (x)")
plt.ylabel("Probabilidad P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Mostrar la gr√°fica
plt.tight_layout()
plt.show()
```

b) ¬øQu√© tipo de sesgo presenta la distribuci√≥n?
La distribuci√≥n no presenta sesgo.
Est√° sim√©trica porque 
ùëù=0.5
p=0.5, lo que implica que los valores alrededor del centro (5) son igualmente probables.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Par√°metros de la distribuci√≥n
n = 10
p = 0.5
mu = n * p  # media

# Valores posibles de X
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

# Gr√°fica de la distribuci√≥n
plt.figure(figsize=(8, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")

# L√≠nea en la media
plt.axvline(mu, color="red", linestyle="--", label=f"Media = {mu}")

# Detalles del gr√°fico
plt.title("Distribuci√≥n Binomial: Simetr√≠a para p = 0.5")
plt.xlabel("x (n√∫mero de caras)")
plt.ylabel("P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

c) Calcula la probabilidad de obtener exactamente 4 caras.

```python
from math import comb

# Par√°metros del problema
n = 10      # n√∫mero de lanzamientos
p = 0.5     # probabilidad de obtener cara
k = 4       # n√∫mero de caras deseado

# C√°lculo de la probabilidad binomial
prob = comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

# Mostrar resultado
print(f"La probabilidad de obtener exactamente {k} caras es: {prob:.5f}")
```

# Ejercicio 8. Piezas defectuosas
 P√°gina 189-190
 Libro: Schaum's Outline of Introduction to Probability and Statistics
 Autores: SEYMOUR LIPSCHUTZ, JOHN J. SCHILLER
 Problema: 
# De una l√≠nea de producci√≥n, se sabe que cada pieza tiene un 2‚ÄØ% de probabilidad de estar defectuosa. Si se inspeccionan 50 piezas, ¬øcu√°l es la probabilidad de encontrar exactamente 2 defectuosas?

a) Determina la distribuci√≥n de probabilidad y traza la gr√°fica correspondiente.
Justificaci√≥n binomial:

Es una distribuci√≥n binomial porque:

Se realiza un n√∫mero fijo de ensayos: 
ùëõ=50
n=50 piezas inspeccionadas.
Cada pieza tiene solo dos resultados posibles: defectuosa (√©xito) o no defectuosa (fracaso).
La probabilidad de √©xito es constante: 
ùëù=0.02
p=0.02.
Se asume que el estado de una pieza es independiente del de las otras.
 Por lo tanto, el experimento cumple con las condiciones para usar la distribuci√≥n binomial.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Par√°metros del ejercicio 3
n = 50          # n√∫mero de piezas
p = 0.02        # probabilidad de que una pieza est√© defectuosa
k = 2           # n√∫mero de piezas defectuosas

# Valores posibles de X
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

# Gr√°fica
plt.figure(figsize=(10, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.bar(k, binom.pmf(k, n, p), color="orange", edgecolor="black", label=f"P(X = {k})")

plt.title("Distribuci√≥n Binomial: Piezas defectuosas (n=50, p=0.02)")
plt.xlabel("N√∫mero de piezas defectuosas (x)")
plt.ylabel("Probabilidad P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

b) ¬øQu√© tipo de sesgo presenta la distribuci√≥n?

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Par√°metros
n = 50
p = 0.02
mu = n * p  # media

# Distribuci√≥n binomial
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

# Graficar
plt.figure(figsize=(10, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.axvline(mu, color="red", linestyle="--", label=f"Media = {mu:.2f}")

plt.title("Distribuci√≥n Binomial: Piezas defectuosas (n=50, p=0.02)")
plt.xlabel("x (n√∫mero de piezas defectuosas)")
plt.ylabel("P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
c) Encuentra la probabilida de encontrar exactamente dos piezas defectuosas
```

```python
from math import comb

# Par√°metros
n = 50       # n√∫mero de piezas
p = 0.02     # probabilidad de que una est√© defectuosa
k = 2        # n√∫mero de defectuosas que se desea encontrar

# C√°lculo de la probabilidad binomial exacta
prob = comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

# Mostrar resultado
print(f"La probabilidad de encontrar exactamente {k} piezas defectuosas es: {prob:.6f}")
```

# Ejercicio 9. Encuesta Politica
P√°gina 204-206
 Libro: Schaum's Outline of Introduction to Probability and Statistics
 Autores: SEYMOUR LIPSCHUTZ, JOHN J. SCHILLER
 Problema: 
 En una encuesta pol√≠tica, la probabilidad de que una persona vote por el candidato A es 0.6. Si se preguntan 20 personas, ¬øqu√© probabilidad hay de que exactamente 15 elijan al candidato A?

a) Determina la distribuci√≥n de probabilidad y traza la gr√°fica correspondiente.
Justificaci√≥n binomial:
La situaci√≥n sigue una distribuci√≥n binomial porque:

Hay un n√∫mero fijo de ensayos: 
ùëõ=20
n=20 personas encuestadas.
Cada persona puede votar por el candidato A (√©xito) o no (fracaso).

La probabilidad de √©xito es constante: 
ùëù=0.6
p=0.6.

Las decisiones de las personas se consideran independientes entre s√≠.

Por lo tanto, se cumple con los 4 criterios de una distribuci√≥n binomial.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Par√°metros del ejercicio 2
n = 20          # n√∫mero de personas encuestadas
p = 0.6         # probabilidad de que voten por A
k = 15          # n√∫mero de personas que votan por A

# Valores posibles de X
x = np.arange(0, n + 1)

# Funci√≥n de masa de probabilidad binomial
y = binom.pmf(x, n, p)

# Gr√°fica
plt.figure(figsize=(8, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.bar(k, binom.pmf(k, n, p), color="orange", edgecolor="black", label=f"P(X = {k})")

# Etiquetas y detalles
plt.title("Distribuci√≥n Binomial: Encuesta pol√≠tica (n=20, p=0.6)")
plt.xlabel("Personas que votan por A (x)")
plt.ylabel("Probabilidad P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

b) ¬øQu√© tipo de sesgo presenta la distribuci√≥n?
Como 
ùëù=0.6>0.5
p=0.6>0.5, la distribuci√≥n binomial est√° ligeramente sesgada a la izquierda. Esto significa que es m√°s probable que ocurra un n√∫mero alto de √©xitos (personas que votan por A), pero hay menos probabilidad en los extremos bajos (0 a 5).



```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Par√°metros
n = 20
p = 0.6
mu = n * p  # media

# Valores posibles de X
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

# Gr√°fica
plt.figure(figsize=(8, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.axvline(mu, color="red", linestyle="--", label=f"Media = {mu:.1f}")

plt.title("Distribuci√≥n Binomial: Encuesta pol√≠tica (n=20, p=0.6)")
plt.xlabel("x (personas que votan por A)")
plt.ylabel("P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

c) Calcular la probabilidad de que exactamente 15 personas voten por el candidato A.

```python
from math import comb

# Par√°metros
n = 20
p = 0.6
k = 15

# Probabilidad exacta con f√≥rmula binomial
prob = comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

# Mostrar resultado
print(f"La probabilidad de que exactamente {k} personas voten por el candidato A es: {prob:.5f}")
```

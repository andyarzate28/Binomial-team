## Ejercicio 1
Pag 164

Seguridad. Seg칰n un anuncio comercial de colchones, 30% de los accidentes automovil칤sticos se
deben a la falta de sue침o. Sup칩n que en un d칤a cualquiera ocurren 10 accidentes automovil칤sticos.


a) Determina la distribuci칩n de probabilidad y traza la gr치fica correspondiente.

Es una distribuci칩n binomial porque:

* Hay un **n칰mero fijo** de accidentes: 10.
* Cada accidente tiene **dos posibles resultados**: se debe o no al sue침o.
* La **probabilidad es constante**: 30%.
* Los accidentes se consideran **independientes** entre s칤.

Por eso, cumple con las condiciones de una distribuci칩n binomial.


```python
n_1 = 10       # n칰mero de accidentes diarios
p_1 = 0.30     # probabilidad de que un accidente se deba al sue침o
k_1 = list(range(n_1 + 1))
y_1 = binom.pmf(k_1, n_1, p_1)
plt.bar(k_1, y_1, color='skyblue', edgecolor='black')
plt.xlabel("N칰mero de accidentes por sue침o (k)")
plt.ylabel("Probabilidad")
plt.title("Distribuci칩n binomial: Accidentes por falta de sue침o")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

b) 쯈u칠 tipo de sesgo presenta la distribuci칩n?

La gr치fica est치 sesgada a la derecha porque hay m치s probabilidad de que ocurran pocos accidentes por sue침o, y menos probabilidad de que ocurran muchos, debido al valor bajo de p=0.30.

c) Calcula la probabilidad de que cinco de los accidentes se hayan debido a la falta de sue침o.

```python
n = 10  # N칰mero total de accidentes
p = 0.30  # Probabilidad de que un accidente sea por falta de sue침o

# Probabilidad de que exactamente 5 accidentes sean por sue침o
prob_5 = binom.pmf(5, n, p)
print(f"Probabilidad de que 5 accidentes sean por sue침o: {prob_5:.4f}")
```

d) Calcula la probabilidad de que al menos dos accidentes hayan sido ocasionados por la falta
de sue침o.

```python
prob_menor_2 = binom.cdf(1, n, p)
prob_al_menos_2 = 1 - prob_menor_2
print(f"Probabilidad de al menos 2 accidentes por sue침o: {prob_al_menos_2:.4f}")
```

e) 쮺u치ntos accidentes esperar칤as que fueran provocados por falta de sue침o?

```python
esperado = n * p
print(f"N칰mero esperado de accidentes por sue침o: {esperado}")
```

Redes sociales. En M칠xico, 6% de los 30 millones de usuarios de internet que hay utilizan Twitter.
(Torres, 2010). Si conoces a 15 personas que sean asiduas a las redes sociales:

a) Determina la distribuci칩n de probabilidad y traza la gr치fica correspondiente.

```python
n = 15       # n칰mero de personas
p = 0.06     # probabilidad de que una persona use Twitter
k = list(range(n + 1))
y = binom.pmf(k, n, p)

plt.bar(k, y, color='mediumpurple', edgecolor='black')
plt.xlabel("N칰mero de personas que usan Twitter (k)")
plt.ylabel("Probabilidad")
plt.title("Distribuci칩n binomial: Uso de Twitter entre 15 personas")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Est치 sesgada hacia la derecha porque es muy poco probable que muchas personas usen Twitter, lo que hace que los valores bajos de k sean los m치s probables.

a) 쮺u치l es la probabilidad de que 9 o m치s utilicen Twitter?

```python
n = 15       # n칰mero de personas
p = 0.06     # probabilidad de que una persona use Twitter
prob_9_o_mas = 1 - binom.cdf(8, n, p)
print(f"Probabilidad de que 9 o m치s personas usen Twitter: {prob_9_o_mas:.10f}")
```

b) 쮺u치l es la probabilidad de que m치ximo dos usen esa red?

```python
prob_max_2 = binom.cdf(2, n, p)
print(f"Probabilidad de que m치ximo 2 personas usen Twitter: {prob_max_2:.4f}")
```

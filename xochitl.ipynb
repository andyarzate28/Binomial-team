## Ejercicio 1
Pag 164

Seguridad. Según un anuncio comercial de colchones, 30% de los accidentes automovilísticos se
deben a la falta de sueño. Supón que en un día cualquiera ocurren 10 accidentes automovilísticos.


a) Determina la distribución de probabilidad y traza la gráfica correspondiente.

Es una distribución binomial porque:

* Hay un **número fijo** de accidentes: 10.
* Cada accidente tiene **dos posibles resultados**: se debe o no al sueño.
* La **probabilidad es constante**: 30%.
* Los accidentes se consideran **independientes** entre sí.

Por eso, cumple con las condiciones de una distribución binomial.


```python
n_1 = 10       # número de accidentes diarios
p_1 = 0.30     # probabilidad de que un accidente se deba al sueño
k_1 = list(range(n_1 + 1))
y_1 = binom.pmf(k_1, n_1, p_1)
plt.bar(k_1, y_1, color='skyblue', edgecolor='black')
plt.xlabel("Número de accidentes por sueño (k)")
plt.ylabel("Probabilidad")
plt.title("Distribución binomial: Accidentes por falta de sueño")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

b) ¿Qué tipo de sesgo presenta la distribución?

La gráfica está sesgada a la derecha porque hay más probabilidad de que ocurran pocos accidentes por sueño, y menos probabilidad de que ocurran muchos, debido al valor bajo de p=0.30.

c) Calcula la probabilidad de que cinco de los accidentes se hayan debido a la falta de sueño.

```python
n = 10  # Número total de accidentes
p = 0.30  # Probabilidad de que un accidente sea por falta de sueño

# Probabilidad de que exactamente 5 accidentes sean por sueño
prob_5 = binom.pmf(5, n, p)
print(f"Probabilidad de que 5 accidentes sean por sueño: {prob_5:.4f}")
```

d) Calcula la probabilidad de que al menos dos accidentes hayan sido ocasionados por la falta
de sueño.

```python
prob_menor_2 = binom.cdf(1, n, p)
prob_al_menos_2 = 1 - prob_menor_2
print(f"Probabilidad de al menos 2 accidentes por sueño: {prob_al_menos_2:.4f}")
```

e) ¿Cuántos accidentes esperarías que fueran provocados por falta de sueño?

```python
esperado = n * p
print(f"Número esperado de accidentes por sueño: {esperado}")
```

## Ejercicio 2
Pag 164

Redes sociales. En México, 6% de los 30 millones de usuarios de internet que hay utilizan Twitter.
(Torres, 2010). Si conoces a 15 personas que sean asiduas a las redes sociales:

a) Determina la distribución de probabilidad y traza la gráfica correspondiente.

```python
n = 15       # número de personas
p = 0.06     # probabilidad de que una persona use Twitter
k = list(range(n + 1))
y = binom.pmf(k, n, p)

plt.bar(k, y, color='mediumpurple', edgecolor='black')
plt.xlabel("Número de personas que usan Twitter (k)")
plt.ylabel("Probabilidad")
plt.title("Distribución binomial: Uso de Twitter entre 15 personas")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Está sesgada hacia la derecha porque es muy poco probable que muchas personas usen Twitter, lo que hace que los valores bajos de k sean los más probables.

a) ¿Cuál es la probabilidad de que 9 o más utilicen Twitter?

```python
n = 15       # número de personas
p = 0.06     # probabilidad de que una persona use Twitter
prob_9_o_mas = 1 - binom.cdf(8, n, p)
print(f"Probabilidad de que 9 o más personas usen Twitter: {prob_9_o_mas:.10f}")
```

b) ¿Cuál es la probabilidad de que máximo dos usen esa red?

```python
prob_max_2 = binom.cdf(2, n, p)
print(f"Probabilidad de que máximo 2 personas usen Twitter: {prob_max_2:.4f}")
```

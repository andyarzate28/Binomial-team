Pag 164

Seguridad. Según un anuncio comercial de colchones, 30% de los accidentes automovilísticos se deben a la falta de sueño. Supón que en un día cualquiera ocurren 10 accidentes automovilísticos.

a) Determina la distribución de probabilidad y traza la gráfica correspondiente.

Es una distribución binomial porque:

Hay un número fijo de accidentes: 10.
Cada accidente tiene dos posibles resultados: se debe o no al sueño.
La probabilidad es constante: 30%.
Los accidentes se consideran independientes entre sí.
Por eso, cumple con las condiciones de una distribución binomial.

```python
n_1 = 10       # número de accidentes diarios
p_1 = 0.30     # probabilidad de que un accidente se deba al sueño
k_1 = list(range(n_1 + 1))
y_1 = binom.pmf(k_1, n_1, p_1)
plt.bar(k_1, y_1, color='skyblue', edgecolor='black')
plt.xlabel("Número de accidentes por sueño (k)")
plt.ylabel("Probabilidad")
plt.title("Distribución binomial: Accidentes por falta de sueño")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

b) ¿Qué tipo de sesgo presenta la distribución?

La gráfica está sesgada a la derecha porque hay más probabilidad de que ocurran pocos accidentes por sueño, y menos probabilidad de que ocurran muchos, debido al valor bajo de p=0.30.

c) Calcula la probabilidad de que cinco de los accidentes se hayan debido a la falta de sueño.

```python
n = 10  # Número total de accidentes
p = 0.30  # Probabilidad de que un accidente sea por falta de sueño

# Probabilidad de que exactamente 5 accidentes sean por sueño
prob_5 = binom.pmf(5, n, p)
print(f"Probabilidad de que 5 accidentes sean por sueño: {prob_5:.4f}")
```

d) Calcula la probabilidad de que al menos dos accidentes hayan sido ocasionados por la falta de sueño.

```python
prob_menor_2 = binom.cdf(1, n, p)
prob_al_menos_2 = 1 - prob_menor_2
print(f"Probabilidad de al menos 2 accidentes por sueño: {prob_al_menos_2:.4f}")
```

e) ¿Cuántos accidentes esperarías que fueran provocados por falta de sueño?

```python
esperado = n * p
print(f"Número esperado de accidentes por sueño: {esperado}")
```

Está sesgada hacia la derecha porque es muy poco probable que muchas personas usen Twitter, lo que hace que los valores bajos de k sean los más probables.

```python
prob_max_2 = binom.cdf(2, n, p)
print(f"Probabilidad de que máximo 2 personas usen Twitter: {prob_max_2:.4f}")
```

Problema: 
a) Determina la distribución de probabilidad y traza la gráfica correspondiente. 

Es una distribución binomial porque: Hay un número fijo de ensayos: 𝑛=10 n=10 lanzamientos de moneda. Cada lanzamiento tiene dos posibles resultados: cara o cruz. La probabilidad de obtener una cara (éxito) es constante: 𝑝=0.5 p=0.5. Los lanzamientos son independientes entre sí.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parámetros del ejercicio
n = 10          # número de ensayos
p = 0.5         # probabilidad de éxito (cara)
k = 4           # número de éxitos deseado

# Valores posibles de X
x = np.arange(0, n + 1)

# Función de masa de probabilidad binomial
y = binom.pmf(x, n, p)

# Graficar la distribución binomial
plt.figure(figsize=(8, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.bar(k, binom.pmf(k, n, p), color="orange", edgecolor="black", label=f"P(X = {k})")

# Personalización del gráfico
plt.title("Distribución Binomial: Lanzamiento de moneda (n=10, p=0.5)")
plt.xlabel("Número de caras (x)")
plt.ylabel("Probabilidad P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

b) ¿Qué tipo de sesgo presenta la distribución?

 La distribución no presenta sesgo. Está simétrica porque 𝑝=0.5 p=0.5, lo que implica que los valores alrededor del centro (5) son igualmente probables.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parámetros de la distribución
n = 10
p = 0.5
mu = n * p  # media

# Valores posibles de X
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

# Gráfica de la distribución
plt.figure(figsize=(8, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")

# Línea en la media
plt.axvline(mu, color="red", linestyle="--", label=f"Media = {mu}")

# Detalles del gráfico
plt.title("Distribución Binomial: Simetría para p = 0.5")
plt.xlabel("x (número de caras)")
plt.ylabel("P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

c) Calcula la probabilidad de obtener exactamente 4 caras.

```python
from math import comb

# Parámetros del problema
n = 10      # número de lanzamientos
p = 0.5     # probabilidad de obtener cara
k = 4       # número de caras deseado

# Cálculo de la probabilidad binomial
prob = comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

# Mostrar resultado
print(f"La probabilidad de obtener exactamente {k} caras es: {prob:.5f}")
```

<span style="color: #153CEE; font-size: 20px; font-weight: bold;">Ejercicio 8</span>.

De una línea de producción, se sabe que cada pieza tiene un 2 % de probabilidad de estar defectuosa. Si se inspeccionan 50 piezas

a) Determina la distribución de probabilidad y traza la gráfica correspondiente.

 Es una distribución binomial porque: Se realizan 𝑛=20 n=20 entrevistas. En cada entrevista, la persona puede votar por el candidato A (éxito) o no (fracaso). La probabilidad de éxito es constante: 𝑝=0.6 p=0.6. Las respuestas de las personas son independientes.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parámetros del ejercicio 3
n = 50          # número de piezas
p = 0.02        # probabilidad de que una pieza esté defectuosa
k = 2           # número de piezas defectuosas

# Valores posibles de X
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

# Gráfica
plt.figure(figsize=(10, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.bar(k, binom.pmf(k, n, p), color="orange", edgecolor="black", label=f"P(X = {k})")

plt.title("Distribución Binomial: Piezas defectuosas (n=50, p=0.02)")
plt.xlabel("Número de piezas defectuosas (x)")
plt.ylabel("Probabilidad P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

b) ¿Qué tipo de sesgo presenta la distribución?

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parámetros
n = 50
p = 0.02
mu = n * p  # media

# Distribución binomial
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

# Graficar
plt.figure(figsize=(10, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.axvline(mu, color="red", linestyle="--", label=f"Media = {mu:.2f}")

plt.title("Distribución Binomial: Piezas defectuosas (n=50, p=0.02)")
plt.xlabel("x (número de piezas defectuosas)")
plt.ylabel("P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

c) Encuentra la probabilida de encontrar exactamente dos piezas defectuosas

```python
from math import comb

# Parámetros
n = 50       # número de piezas
p = 0.02     # probabilidad de que una esté defectuosa
k = 2        # número de defectuosas que se desea encontrar

# Cálculo de la probabilidad binomial exacta
prob = comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

# Mostrar resultado
print(f"La probabilidad de encontrar exactamente {k} piezas defectuosas es: {prob:.6f}")
```

Problema: En una encuesta política, la probabilidad de que una persona vote por el candidato A es 0.6. Si se preguntan 20 personas

a) Determina la distribución de probabilidad y traza la gráfica correspondiente. Justificación binomial:

 La situación sigue una distribución binomial porque:

Hay un número fijo de ensayos: 𝑛=20 n=20 personas encuestadas. Cada persona puede votar por el candidato A (éxito) o no (fracaso).

La probabilidad de éxito es constante: 𝑝=0.6 p=0.6.

Las decisiones de las personas se consideran independientes entre sí.

Por lo tanto, se cumple con los 4 criterios de una distribución binomial.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parámetros del ejercicio 2
n = 20          # número de personas encuestadas
p = 0.6         # probabilidad de que voten por A
k = 15          # número de personas que votan por A

# Valores posibles de X
x = np.arange(0, n + 1)

# Función de masa de probabilidad binomial
y = binom.pmf(x, n, p)

# Gráfica
plt.figure(figsize=(8, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.bar(k, binom.pmf(k, n, p), color="orange", edgecolor="black", label=f"P(X = {k})")

# Etiquetas y detalles
plt.title("Distribución Binomial: Encuesta política (n=20, p=0.6)")
plt.xlabel("Personas que votan por A (x)")
plt.ylabel("Probabilidad P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

b) ¿Qué tipo de sesgo presenta la distribución? 


Como 𝑝=0.6>0.5 p=0.6>0.5, la distribución binomial está ligeramente sesgada a la izquierda. Esto significa que es más probable que ocurra un número alto de éxitos (personas que votan por A), pero hay menos probabilidad en los extremos bajos (0 a 5).

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parámetros
n = 20
p = 0.6
mu = n * p  # media

# Valores posibles de X
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

# Gráfica
plt.figure(figsize=(8, 5))
plt.bar(x, y, color="skyblue", edgecolor="black")
plt.axvline(mu, color="red", linestyle="--", label=f"Media = {mu:.1f}")

plt.title("Distribución Binomial: Encuesta política (n=20, p=0.6)")
plt.xlabel("x (personas que votan por A)")
plt.ylabel("P(X = x)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

c) Calcular la probabilidad de que exactamente 15 personas voten por el candidato A.

```python
from math import comb

# Parámetros
n = 20
p = 0.6
k = 15

# Probabilidad exacta con fórmula binomial
prob = comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

# Mostrar resultado
print(f"La probabilidad de que exactamente {k} personas voten por el candidato A es: {prob:.5f}")
```
